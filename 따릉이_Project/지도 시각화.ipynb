```python
import pandas as pd
import numpy as np
import folium
```

```python
food_df= pd.read_csv("data/food_data.csv", encoding='EUC-KR', index_col=0) 
bike_df= pd.read_csv("data/bike_data_Gangnam.csv", encoding='cp949', index_col=0) 
```

```python
# 위도 경도 값 활용해 지도 시각화
food_df[['업소명', '위도', '경도']].head()
```

```python
food_df.shape
```

```python
# 지도의 중심 지정: 전체 맛집 위치의 위도 경도 평균
lat= food_df['위도'].mean()
long= food_df['경도'].mean()
m= folium.Map([lat, long], zoom_start=13)       # zoom 으로 축적 설정

# 지도 위에 마커로 표시
for i in food_df.index:
    lat2= food_df.loc[i, '위도']
    long2= food_df.loc[i, '경도']
    title= food_df.loc[i, '업소명']
    folium.Marker([lat2, long2], tooltip= title).add_to(m)
m
```

```python
# 지도 위에 마커로 표시
for i in bike_df.index:
    lat3= bike_df.loc[i, '위도']
    long3= bike_df.loc[i, '경도']
    title= bike_df.loc[i, '대여소 이름']
    folium.Marker([lat3, long3], tooltip= title, icon = folium.Icon(color='green')).add_to(m)
m
```
