# place

```python
place = pd.read_csv("place_data_평점포함.csv", encoding = 'cp949')
place.head()
```

```python
place["분류"].value_counts()
```

* 관광지별 소요시간 정하기
- 쇼핑 : 1시간 30분
- 문화관광 : 3시간
- 역사관광지 : 2시간 30분
- 휴양관광지 : 2시간

```python
place["소요시간"] = np.where(
    place["분류"] == "문화관광", 3, np.where(
        place["분류"]=="쇼핑", 1.5, np.where(
            place["분류"] == "역사관광지", 2.5 , 2)))
```
