```python
import pandas as pd
import numpy as np
```

```python
food_df= pd.read_csv("data/20220914163333_지역_맛집_외지인.csv", encoding='EUC-KR', index_col=0) 
```

```python
food_df
```

```python
# 결측값 개수 구하기
missing= food_df.shape[0] - food_df.count()
print(missing)
```

```python
f_add = food_df['주소']
for i in range(1, 101):
    if str(f_add[i][-2:]) == "-0":
        f_add[i]=f_add[i].strip("-0")
print(f_add)
```

```python
# 위도 경도 추출 위해 geopy 라이브러리 import
import geopy
from geopy.geocoders import Nominatim
geo_local= Nominatim(user_agent='South Korea')
# user_agent='South Korea': 한글주소를 지오코딩 할 수 있음
```

```python
# 위도, 경도 값 반환하는 함수 생성
def geocoding(f_add):
    geo= geo_local.geocode(f_add)
    x_y= [geo.latitude, geo.longitude]  # 주소에 맞는 위도, 경도를 [x, y] 형태의 리스트로 반환해라
    return x_y
```

```python
# 위도, 경도 값 추출
latitude= []
longitude= []

for i in f_add:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
# 데이터프레임 만들기
food_df_geo= pd.DataFrame({'업소명': food_df['업소명'],
                   '주소': food_df['주소'],
                   '위도': latitude,
                   '경도': longitude,
                   '분류': food_df['분류']})
food_df_geo= food_df_geo.reset_index(drop=True)
food_df_geo
```

```python
# 새로 가공한 맛집 데이터를 csv로 저장
food_df_geo.to_csv('data/food_data.csv', encoding='EUC-KR')
```
